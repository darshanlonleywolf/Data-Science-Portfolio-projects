```python
from transformers import pipeline

sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
print(sentiment_analysis("its good but still bad"))
```

```python
kc_df['sentiment analysis']=kc_df['content'].apply(lambda x: sentiment_analysis(x))

kc_df.sample(7)
```
